```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/cat_image.jpg")
# cv2.imshow("CAT", image)
cv2_imshow(image)
```

```python
print(image.shape)
# image[height][width][channel]
```

```python
print(image)
# image - 3D ARRAY (TENSOR)
# grayscale image - 2D ARRAY (MATRIX)
```

```python
# channel: [B, G, R] = [127 128 138]
(B, G, R) = image[100, 100]
print(f"Blue: {B}, Green: {G}, Red: {R}")

# OpenCV  BGR (Windows bitmap BMP)
# PIL | matplotlib | YOLO | PyTorch  RGB
new_image = image.transpose(2,0,1)
```

# **GRAYSCALE**

```python
grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # 3D TENSOR > 2D MARTIX
# print(grayscale)
cv2_imshow(grayscale)
print(grayscale.shape)

# image[height][width][channel] > image[height][width]
# GRAY = 0.299 x RED + 0.587 x GREEN + 0.114 x BLUE
```

```python
resized = cv2.resize(grayscale, (300, 300))
cv2_imshow(resized)
```

```python
cropped = grayscale[100:400, 550:800] # [Y1:Y2, X1:X2]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(cropped, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(rotated, 1)
cv2_imshow(flipped)
```

# **SAVE**

```python
cv2.imwrite("crf_cat.jpg", flipped)
```

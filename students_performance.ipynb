```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.tail()
```

# **MALUMOTLAR BILAN TANISHUV**

```python
print(f"DataSet shape (row, column): {df.shape} \n")
print(f"Columns name: {df.columns.tolist()} \n")

print("✅ Top 5 data:\n", df.head(), "\n")

print("✅ Summary statistics:\n", df.describe(), "\n")

print("✅ Data types & null info:")
print(df.info())
```

# **STATISTIK AMALIY AMALIYOT**

```python
print("✅ Average score by columns:")
df[["math score", "reading score","writing score"]].mean()
```

```python
print("✅ Max score by columns:")
df[["math score", "reading score","writing score"]].max()
```

```python
print("✅ Min score by columns:")
df[["math score", "reading score","writing score"]].min()
```

```python
print("✅ Correlation:")
df[["math score", "reading score","writing score"]].corr()
```

```python
df["total score"] = df[["math score", "reading score","writing score"]].sum(axis=1) # row yani qator boyicha yig'amiz
df["average score"] = df["total score"] / 3
df["passed"] = np.where(df["average score"] >= 50, "Yes", "No")

print("✅ AMALIY TAHLIL:")
df[["total score", "average score", "passed"]].head(10)
```

# **GROUP BASED ANALYSIS**

```python
print("✅ Guruh boyicha tahlil:")
test_group_analysis = df.groupby("test preparation course")[["math score", "reading score","writing score"]].mean()
print(test_group_analysis, "\n")
# Xulosa: test preparation course olganlar yuqori natija qayd etgan!

print("✅ Gender boyicha tahlil:")
gender_analysis = df.groupby("gender")[["math score", "reading score","writing score"]].mean()
print(gender_analysis, "\n")
# Xulosa: Erkaklar matematikada kuchli, ayollar esa reading va writingda kuchli ekan!

print("✅ Muvaffaqiyat boyicha tahlil:")
passed_analysis = df["passed"].value_counts(normalize=True) * 100
print(passed_analysis, "\n")
```

```python
# values state property
scores = df[["math score", "reading score","writing score"]].values

# Mean
print("✅ NumPy mean:\n", np.mean(scores, axis=0), "\n")

# Std
std_score = np.std(scores, axis=0)
print("✅ NumPy std:\n", std_score , "\n")

# Z-score (normalize)
z_score = (scores - np.mean(scores, axis=0)) / std_score
print("✅ z_score:\n", z_score[0:5]) # Average 0 | Positive + |Negative -
```

```python
top_scored = df[df["average score"] >= 90 ]
print("✅ Top scored:\n", top_scored["average score"].head() , "\n")
print("✅ Top scored counts:", len(top_scored) , "\n")

failed_students = df[df["passed"] == "No" ]
print("✅ Failed counts:", len(failed_students) , "\n")
```

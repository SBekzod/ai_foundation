```python
# Dictionary
mydataset = {
  'cars': ["BMW", "Volvo", "Ford", "Tesla", "Ferrari", "Bugatti", "BMW"],
  'counts': [3, 7, 2, 5, 8, 14, 17]
}
print(mydataset, "\n")

df = pd.DataFrame(mydataset)
print(df, "\n")


```

```python
# head method
df.head(4) # default 5 rows
```

```python
# tail method
df.tail(2) # default 5 rows
```

```python
# State Property: shape, column, dtypes
a = df.shape
print("=====\n", a)

b = df.columns
print("=====\n", b)

c = df.dtypes
print("=====\n", c)

df["cars"] = df["cars"].astype("string")
d = df.dtypes
print("=====\n", d)
```

```python
# describe method - numeric statistikani hosil qiladi!
df.describe()
# df.describe(include="all")
```

```python
# iloc (integer location) method - kerakli malumotlarini tanlab olish

print("=====\n", df.iloc[0])

print("=====\n", df.iloc[1])

print("=====\n", df.iloc[1,0]) # [row location, column location]

print("\n")
df.iloc[1:4]


```

```python
# loc method - kerakli malumotlarni ma'lum bir columndan tanlab olish
df.loc[1:4, "cars"]
```

**FILTERING**

```python
dataset = {
  'cars': ["BMW", "Volvo", "Ford", None, "Ferrari", "Bugatti", None],
  'counts': [3, 7, 2, None, 8, 14, 17]
}

df_missing = pd.DataFrame(dataset)
print(df_missing, "\n")
```

```python
# None nisbattan tekshiramiz
df_missing.isnull()
```

```python
# None holatdagi malumotlarni aniqlaymiz
df_missing.isnull().sum()
```

```python
# dropna method - Missing malumotlardan tozalaymiz

df_clean = df_missing.dropna()
print(df_clean)
```

```python
# df_missing.dropna(how="all") # har ikkalasi None bolganini ochiradi
df_missing.dropna(axis=0) # tanlangan 0-row 1-column None boyicha ochiradi
# df_missing.dropna(subset="counts") # tanlangan column None tozalaydi
```

```python
# fillna method - ma'lumotlarni tog'irlab toldirib olish

# df_missing.fillna("unknown")
df_missing.fillna({"cars": "unknown", "counts": 0})
```

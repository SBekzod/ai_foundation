```python
import matplotlib.pyplot as plt
import numpy as np
```

LINE GRAPH

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 40])

plt.plot(x, y)
plt.title("Basic Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 2, 10])

plt.plot(
    x, y, linestyle="dotted",
    linewidth=3, color="orange",
    markerfacecolor="green",
    markeredgecolor="blue", marker="o"
)
plt.title("Styled Line")
plt.show()
```

MULTI LINE GRAPH

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 5, 5])
y2 = np.array([1, 3, 5, 7])

plt.plot(x, y1, label="Line 1", color="red", marker="s")
plt.plot(
    x, y2, label="Line 2",
    color="blue", linestyle="--", marker="o"
)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

```

BAR CHART

```python
x = np.array(["Salary", "Expense", "Charity"])
y = np.array([500, 300, 30])

plt.bar(x, y, color="orange")
plt.title("Bar char money")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

SCATTER PLOT

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 2, 10])

plt.scatter(x, y, color="purple", s=150)
plt.title("Scatter plot")
plt.show()
```

HISTOGRAM

```python
data = np.array([1, 2, 3, 4, 3, 8, 2, 10])

plt.style.use("ggplot")
plt.hist(
    data, bins=5, color="green",
    edgecolor="black", linewidth=1.5, alpha=0.7
)

plt.title("Histogram", fontsize=16, fontweight="bold")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

Subplot (Multiple Plots together)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')

plt.subplot(1, 2, 1)
plt.scatter(x, y2, color='green')

plt.tight_layout()
plt.savefig("subplot.png")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
